```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime
```

```python
columns = ["Ticker", "Date", "Open", "High", "Low", "Close", "Volume"]

df = pd.read_csv("/content/Stock_Price_Dataset.csv", names=columns, header=None)

df = df[df["Ticker"].str.len() <= 5]

df.head()
```

```python
df = df.iloc[1:].reset_index(drop=True)

df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

ups_df = df[df['Ticker'] == 'UPS']

plt.figure(figsize=(10, 5))
plt.plot(ups_df['Date'], ups_df['Close'], color='blue', marker='o')
plt.title("UPS Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.isnull().sum()
```

```python
uri_df = df[df['Ticker'] == 'URI']

plt.figure(figsize=(10, 5))
plt.plot(uri_df['Date'], uri_df['Close'], color='green', marker='o')
plt.title("URI Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```
